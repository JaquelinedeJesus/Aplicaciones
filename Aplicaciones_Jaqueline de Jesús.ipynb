```python
from sympy.core import symbol
from sympy import *
from math import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("Durante una ronda de inspección por una reserva, dos guardabosques descubrieron el cadáver de un jabalí. Un examen preliminar permitió concluir"
 " que el animal falleció instantáneamente a causa de un balazo de un cazador furtivo. Sospechando que el cazador debía regresar a recoger su presa,"
  "los guardas decidieron esperarlo escondidos no lejos del lugar del crimen. Al poco tiempo aparecieron dos sujetos y se dirigieron sin rodeos hada el jabalí."
      "Al ser detenidos, los desconocidos negaron rotundamente su participación en el delito. Aunque los guardabosques tenían pruebas indirectas de su "
      "culpabilidad, para obtener una prueba fehaciente era necesario determinar el instante exacto en que el jabalí fue muerto.Por fortuna, ese instante se "
      "logró calcular utilizando la ley de radiación de calor. Conforme a ley de radiación"
      "de calor, la velocidad de enfriamiento de un cuerpo en el medio ambiente es proporcional a la diferencia entre la temperatura del cuerpo y la temperatura del medio:")
print('dx/dt=-K(x-a) con x=temperatura, t=tiempo, k=constante, a=temperatura del ambiente')
print('SOLUCIÓN')
print('Existen dos posibles casos:')
print('Caso 1: a=constante')
print('Resolvemos por variables separadas')
print('d(x)/d(x)=-k*(x-a)')
print('log(x-a/x0-a)=(-k)*(t) con x ≠ a')
print('sustituimos con x=37°C, a=21°C, x0= 29°C')
k=log((31-21)/(29-21))
print('k=',k)
print('Ahora, sustituimos con k=0.22314 y x=37')
t=(-1/k)*(log((37-21)/(31-21)))
print('t=',t)
print('Esto indica que entre el momento del disparo y el momento de la detención, pasaron 2 horas con 6 min')
print('Caso 2: a=a(t)')
print('Resolvemos por factor integrante con a(t)=t*-t y t*=instante en que los guardabosques hallaron los cádaveres')
print('d(x)/d(t)+kx=-k*a(t)')
print('Ahora, suponemos lo siguiente: al momento de la detención t=30°c, ese día la temperatura después del medio dia aumentaba 1°C cada hora y era igual a 0°C cuando se halló el cádaver y la temperatura del animal después de una hora de ser detetctado bajó a 25°C. Tomando t=0 y x0=37°C')
print('Integramos:')
print('x=(37-t*-(1/k))*(e^-kt)+t*-t+(1/k)')
print('Teniendo en cuenta x = 30 para t=t* y  x=25 para t = t*+1, de la última expresión obtenemos ')
print('(37-t*-(1/k))*(e^-kt*)+(1/k)=30')
print('(37-t*-(1/k))*(e^-kt*)+(1/k)=26')
print('partiendo de ellas, tenemos:')
print('30-(1/k)*(e^-k)-26+(1/k)=0')
print('ahora, para poder usar el método de Newton, transformamos las ecuaciones anteriormente obtenidas')
print('30k-1+(1-26k)*e^k=0')
print('(37k-1+kt)*(e^kt)-30k+1=0')
print('Notamos que estas dos últimas ecuaciones son del tipo: (ax+b)*(e^λx)+cx+d=0, por lo que denotamos al primer miembro de la ecuación y derivamos dos veces con respecto de x:')
print("p'(x)=(ax+λb+a)*(e^λx)+c")
print("p''(x)=(λ^2ax+λ^2b+2λa)*(e^λx)")
print('Apliquemos el método de Newton  y calculemos una raíz con una precisión de 10^-6 . Derivando el primer miembro p(k) respecto a k:')
print("p'(k)=30-(25-26k)*(e^k) con un intervalo (0,1) y k0=0.5")
print('Usamos Newton')
fx = lambda x: 30*x-1+(1-26*x)*(2.718281828**x)
dx = lambda x: 30-(24-26*x)*(2.718281828**x)
xi=0.5
columnas=["iteración","xi","f(xi)","f'(xi)","xi+1","Error"]
datos=[[0,0,0,0,xi,0]]
df = pd.DataFrame(datos,columns=columnas)

i=1
tolerancia=0.00001
error=1
while error>tolerancia:
    Raiz=xi-(fx(xi)/dx(xi))
    error = np.abs((Raiz-xi)/Raiz)
    df = df.append({
      "iteración":1,
      "xi":xi,
      "f(xi)":fx(xi),
      "f(xi)":dx(xi),
      "xi+1":Raiz,
      "Error":error
    },ignore_index=True)
    xi=Raiz
    i+=1
print("La raiz es: ",Raiz)
print("El valor de f(x): ",fx(Raiz))
df.head(15)
xs=np.linspace(Raiz-1,Raiz+1,50)

plt.plot(xs,fx(xs),label="f(x)")
plt.title("Método de Newton Raphson")
plt.axhline(y=0,color="blue")
plt.plot(Raiz,fx(Raiz),"bo",label= f'f(x)=0,x=(Raiz:6f)',color='k')
plt.legend(loc='upper left')
df.head(i)
print('ahora hacemos lo mismo pero para k0=0.178952')
fx = lambda x: 30*x-1+(1-26*x)*(2.718281828**x)
dx = lambda x: 30-(24-26*x)*(2.718281828**x)
xi=0.178952
columnas=["iteración","xi","f(xi)","f'(xi)","xi+1","Error"]
datos=[[0,0,0,0,xi,0]]
df = pd.DataFrame(datos,columns=columnas)

i=1
tolerancia=0.00001
error=1
while error>tolerancia:
    Raiz=xi-(fx(xi)/dx(xi))
    error = np.abs((Raiz-xi)/Raiz)
    df = df.append({
      "iteración":1,
      "xi":xi,
      "f(xi)":fx(xi),
      "f(xi)":dx(xi),
      "xi+1":Raiz,
      "Error":error
    },ignore_index=True)
    xi=Raiz
    i+=1
print("La raiz es: ",Raiz)
print("El valor de f(x): ",fx(Raiz))
df.head(15)
xs=np.linspace(Raiz-1,Raiz+1,50)

plt.plot(xs,fx(xs),label="f(x)")
plt.title("Método de Newton Raphson")
plt.axhline(y=0,color="blue")
plt.plot(Raiz,fx(Raiz),"bo",label= f'f(x)=0,x=(Raiz:6f)',color='k')
plt.legend(loc='upper left')
df.head(i)
print('El jabalí fue cazado 1 hr y 12 min antes de que lo encontrarán')
```
